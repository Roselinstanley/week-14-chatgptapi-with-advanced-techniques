```python
def prompt_response_tokencount(messages):
    prompt_response=client.models.generate_content(model="gemini-2.5-flash",
                                                   contents=f"Respond in two lines.\n\n{messages}",
                                                   config={
                                                                       "temperature":1,
                                                                       "max_output_tokens":500}
                                                  )
    prompt_response_content=prompt_response.text
    print(prompt_response_content)
    usage=prompt_response.usage_metadata
    return prompt_response_content,usage
```
